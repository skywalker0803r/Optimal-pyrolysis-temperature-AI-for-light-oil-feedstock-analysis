```python
import pandas as pd
import os
from tqdm import tqdm_notebook as tqdm
pd.options.display.max_rows = 999

def preprocess1(df):
    df = df.T
    df = df.iloc[:,[1,2,3,5,7,8,21,22]]
    df.columns = df.iloc[1,:-2].values.tolist() + ['乙烯產量','丙烯產量']
    df.index.name='日期'
    df = df.iloc[2:-2,:]
    return df.dropna(axis=0)


root = '../data/OL_1資料/'
file_list = os.listdir(root)
df = pd.DataFrame()
for file_name in tqdm(file_list[:]):
    try:
        d = pd.read_excel(root+file_name)
        d = preprocess1(d)
        df = df.append(d)
    except:
        print(file_name,'有問題')

display(df.head())
df.to_csv('../data/cleaned_PONA_data.csv')
```
