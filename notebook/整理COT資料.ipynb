```python
import pandas as pd

path1 = '../data/(數值檔)輕油原料COT分析2018-2021年 ver3.xlsx'
path2 = '../data/(數值檔)輕油原料COT分析 2020年.xlsx'

def process_cot_data(path):
    df = pd.read_excel(path,index_col=0)
    df.index.name = '日期'
    #v = ['H-'+str(i)+' COT' for i in range(110,200,10)]
    v = df.columns
    df = df.loc[:,v].iloc[1:]
    return df
```

```python
df1 = process_cot_data(path1)
df1
```

```python
c1 = ['H-'+str(i)+' COT' for i in range(110,200,10)]
c2 = df1.columns[df1.columns.str.contains('裂解度')].tolist()
df = df1[c1+c2]
df.head()
```

```python
df.to_csv('../data/cleaned_COT_data.csv')
df = pd.read_csv('../data/cleaned_COT_data.csv')
print(df.shape)
df.head()
```
